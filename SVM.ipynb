```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
df = load_breast_cancer()
```

# loading datasetsa also taking feature and labels 

```python
x,y = df.data,df.target
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x = scaler.fit_transform(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train_set,x_test_set,y_train_set,y_test_set = train_test_split(x,y,test_size=0.2,random_state=42)
```

## Train SVM Models

```python
from sklearn.svm import SVC
```

```python
#linear kernel
svm_linear = SVC(kernel='linear',C=1)
svm_linear.fit(x_train_set,y_train_set)

#RBF kernel
svm_rbf = SVC(kernel="rbf",C=1,gamma=0.1)
svm_rbf.fit(x_train_set,y_train_set)
```

##  Visualize Decision Boundary

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons

# Generate toy dataset
X2, y2 = make_moons(n_samples=200, noise=0.2, random_state=42)

# Train SVM
svm = SVC(kernel="rbf", C=1, gamma=0.5)
svm.fit(X2, y2)

# Plot decision boundary
import numpy as np
xx, yy = np.meshgrid(np.linspace(X2[:,0].min()-1, X2[:,0].max()+1, 100),
                     np.linspace(X2[:,1].min()-1, X2[:,1].max()+1, 100))
Z = svm.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.2)
plt.scatter(X2[:,0], X2[:,1], c=y2, edgecolors='k')
plt.title("SVM Decision Boundary (RBF Kernel)")
plt.show()
```

## Tune Hyperparametors


```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'C': [0.1, 1, 10],
    'gamma': [0.01, 0.1, 1],
    'kernel': ['rbf']
}

grid = GridSearchCV(SVC(), param_grid, cv=5)
grid.fit(x_train_set, y_train_set)

print("Best Params:", grid.best_params_)
print("Best Score:", grid.best_score_)
```

## Using Cross Validation

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(SVC(kernel="rbf", C=1, gamma=0.1), x, y, cv=5)
print("Cross-validation accuracy:", scores.mean())
```
